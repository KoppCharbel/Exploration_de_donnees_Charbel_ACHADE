```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# Charger le fichier csv "201902_fordgobike-tripdata"

df = pd.read_csv('201902_fordgobike-tripdata.csv')
```

*Notez que les cellules ci-dessus ont été définies comme des diapositives de type "Passer". Cela signifie que lorsque le bloc-notes est rendu sous forme de diapositives http, ces cellules ne s'affichent pas.*

## Visualisation 1

*Je me suis concentré dessus pour connaître la moyenne. Dans cet ensemble de données, la durée était en secondes. Lorsque j'ai essayé de faire une exploration univariée, cela m'a montré une visualisation difficile à interpréter. J'ai dû convertir la durée en secondes en minutes. La visualisation que j'ai obtenue est plus nette.*

```python
# la durée des trajets par seconde  

plt.figure(figsize=(10,6))
bins = np.arange(0, 10000, 50)


plt.hist(data=df, x='duration_sec', bins=bins)
plt.xlabel('Durée de Trajet')
plt.ylabel('Le Nombre de Personnes ')
plt.title("Le temps en seconde par nombre d'utilisateur des trajets ");
```

```python
# Changer la duree en minutes 

df['duration_min'] = df['duration_sec']/60

df.duration_min = df.duration_min.astype('int64')
```

```python
# la durée des trajets par minute 

plt.figure(figsize = (10,10))

df['duration_min'].hist(bins=60)

plt.xlim(0, 60)
plt.xlabel('Durée (mn)')
plt.ylabel("Nombre d'utilisateurs")
plt.title("Durée des trajets par Minute")

plt.show()
```

## Visualisation 2

*J'ai trouvé ces données très importantes pour ce service car il est essentiel de savoir quelles personnes utilisent le service. Pour le savoir, j'ai utilisé un "countplot" avec le type d'utilisateur comme variable. Il m'a montré qu'environ 160 000 personnes sont abonnées et moins de 20 000 personnes sont clientes.*

```python
# Nombre d'utilisateurs par type de user 

coul_base = sb.color_palette()[0] # Defnir la couleur de base (bleue)

plt.figure(figsize=[7, 4])

sb.countplot(data = df, x = 'user_type', color = coul_base) # Fonction qui permet de tracé le plot

plt.title("Tracé du nombre d'utilisateurs par type de user ") # Definir le titre, le nom sur l'axe x et y 
plt.ylabel("Nombre d'utilisateurs")
plt.xlabel("Different Type d'utilisateurs");
```

## Visualisation 3

*Pour obtenir ce resultat, j'ai utilisé une visualisation bivariée. Elle me permet de savoir quelle est la relation entre le sexe du membre et le type d'utilisateur. De cela nous pouvons deduire que pour chaque sexe de membre, il y a plus d'homme que de femme.*


```python
# Code pour visualiser la relation entre le sexe du membre et le type d'utilisateur

plt.figure(figsize = [7, 4]);

sb.countplot(data = df, x = 'member_gender', hue = 'user_type');

plt.title("Répartition entre le sexe du membre et le type d'utilisateur");
```
